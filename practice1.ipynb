```python
#q10
df = pd.read_csv('data_q10.csv')

df
```

```python
#q10_1

print("الف: \n")

print("mean mark: ",df['نمره'].mean())
print("max mark: ",df['نمره'].max())

m_mean= df[df['جنسیت']=='M']['نمره'].mean()
f_mean=df[df['جنسیت']=='F']['نمره'].mean()

print("\nmen's mean was: ",m_mean)
print("women's mean was: ",f_mean)

if m_mean>f_mean:
  print("\nthe highest mean was for Male")
else:
  print("\nthe highest mean was for Female")
```

```python
#q10_2

m_std= df[df['جنسیت']=='M']['نمره'].std()
f_std= df[df['جنسیت']=='F']['نمره'].std()

plt.bar(['Male','Female'],[m_mean,f_mean],yerr=[m_std,f_std])
plt.xlabel('Gender')
plt.ylabel('Marks mean')
plt.title('mean and std of gender diagram')
plt.show()
```
